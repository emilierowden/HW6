b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. ```python
#your code goes here
import matplotlib.pyplot as plt
plots = sns.relplot(data=house_states, x="date", y="house_price_index", kind="line", color="red")
plots.set_xticklabels(['1980', '1980', '2000', '2000', '2020', "2020"])
plt.xlabel('')
plt.ylabel('house price index')
```

c) Draw a lineplot for 'house price perc', but color them based on the 'state' name which will have four lines in a single plot. ```python
#your code goes here
sns.lineplot(data=house_states, x="date", y="house_price_perc", palette=['r', 'g'])
plt.legend(loc='lower center', labels=[states])
plt.xlabel('')
plt.ylabel('house price perc')
```

```python
#A
team_pairs = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).count().reset_index()
team_pairs.head()
```

Make sure that you only write four columns ('Home Team Initials', 'Away Team Initials', 'weight', 'HomeGoalTotal') from the data frame, team_pairs, to a csv file

```python
#B
home_goal_total= world_cup.groupby(['Home Team Initials','Away Team Initials']).sum(['Home Team Goals']).reset_index()
home_goal_total.head()
```

```python
#C
from google.colab import files
team_pairs = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).count().reset_index()
team_pairs['Home Goal Sums'] = home_goal_total['Home Team Goals']
team_pairs.head()

team_pairs.to_csv('team_pairs.csv') 
files.download('team_pairs.csv')
```
